```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from PIL import Image
```

```python
# Step 1: Load and review the image
image_path = 'New York.jpg'
image = Image.open(image_path)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Step 2: Reshape the image for k-means clustering
image_array = np.array(image)
height, width, channels = image_array.shape
reshaped_image = image_array.reshape(-1, channels)
```

```python
# Step 3: Run k-means clustering on reshaped image
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(reshaped_image)
```

```python
# Step 4: Print cluster centers
cluster_centers = kmeans.cluster_centers_
print("Cluster Centers:")
print(cluster_centers)
```

```python
# Step 5: Construct compressed image
compressed_labels = kmeans.labels_
compressed_image = cluster_centers[compressed_labels].astype('uint8')
compressed_image = compressed_image.reshape(height, width, channels)
```

```python
# Step 6: Show the compressed image
plt.imshow(compressed_image)
plt.title('Compressed Image')
plt.axis('off')
plt.show()
```
